```python
import pandas as pd
import json
```

```python
with open('./file1.json', 'r') as f:
    json_data = json.load(f)
```

```python
# json_data[1]
```

```python
filtered = list(filter(lambda x: 'restaurants' in x, json_data))
```

```python
restaurants = list(map(lambda x: x['restaurants'], filtered))
```

```python
from functools import reduce

restaurants_reduced = list(reduce(lambda x, y: x + y, restaurants))
```

```python
restaurant_data = list(map(lambda x: x['restaurant'], restaurants_reduced))

normalized = pd.json_normalize(restaurant_data)
```

```python
df = pd.DataFrame.from_dict(normalized)

df
```

```python
df.iloc[0]
```

```python
df_filtered = df[['has_online_delivery', 'price_range', 'currency', 'name', 'cuisines', 'location.address', 'location.city', 'user_rating.rating_text']]
```
