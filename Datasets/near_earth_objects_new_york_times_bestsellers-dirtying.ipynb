```python
import pandas as pd
import requests
import time
import numpy as np
```

```python
endpoint = 'https://api.nasa.gov/neo/rest/v1/feed?start_date=1995-01-01&end-date=1995-01-08&api_key=KidBlDQkkXeu9dFpdV5GSVDrTH2H640zggDaZJ1a'
```

```python
dict_datos = {}

for i in range(0, 8):
    
    try:
        time.sleep(1)
        r = requests.get(endpoint, timeout=5)

        if r.status_code == 200:
            json = r.json()
            endpoint = json['links']['next']
            dict_datos[i] = json['near_earth_objects']
    except Exception as e:
        print(e)
        continue
```

```python
date_to_dataframe = {}

for key, value in dict_datos.items():
    for key_2, value_2 in value.items():
        normalized = pd.json_normalize(value_2)
        df = pd.DataFrame.from_dict(normalized)
        date_to_dataframe[key_2] = df.reset_index(drop=True)

    
```

```python
dates = date_to_dataframe.keys()
dataframes = date_to_dataframe.values()

df = pd.concat(dataframes, axis=0).reset_index(drop=True)

df
```
